```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import export_graphviz
import graphviz

# 直接构建数据
data = {
    'age': ['<=30', '<=30', '30..40', '>40', '>40', '>40', '31..40', '<=30', '<=30', '>40', '<=30', '31..40', '31..40', '>40'],
    'income': ['high', 'high', 'high','medium', 'low', 'low', 'low','medium', 'low','medium','medium','medium', 'high','medium'],
   'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}
data = pd.DataFrame(data)

# 特征列
X = data[['age', 'income','student', 'credit_rating']]
# 目标列
Y = data['buys_computer']

# 对类别型特征进行编码
for col in X.columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
le_Y = LabelEncoder()
Y = le_Y.fit_transform(Y)

# 划分训练集和测试集
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 创建基于ID3近似（entropy准则）的决策树模型
dtree_id3 = DecisionTreeClassifier(criterion='entropy')
# 训练模型
dtree_id3.fit(X_train, Y_train)
# 预测
Y_pred = dtree_id3.predict(X_test)
# 评估模型
accuracy = accuracy_score(Y_test, Y_pred)
print("基于ID3近似的决策树模型在测试集上的准确率:", accuracy)

# 可视化决策树
dot_data = export_graphviz(dtree_id3, out_file=None,
                           feature_names=X.columns,
                           class_names=le_Y.classes_,
                           filled=True, rounded=True,
                           special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
